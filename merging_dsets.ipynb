# a custom example

```python
A = pd.DataFrame({
    'Fruit': ['apple', 'graips', 'pear', 'apricot'],
    'Supply': [30, 20, 10, 20]
})
A
```

```python
B = pd.DataFrame({
    'Fruit': ['apple', 'graips', 'banana', 'berry'],
    'Price': [8000, 20_000, 24_000, 18_000]
})
B
```

```python
C = pd.merge(A, B, on = 'Fruit', how = 'inner')
C
```

# Outer merge

```python
C1 = pd.merge(B, A, on = 'Fruit', how = 'outer')
C1
```

# merging based on specific columns

```python
A1 = pd.DataFrame({
    'Fruit_name': ['apple', 'graips', 'pear', 'apricot'],
    'Supply': [30, 20, 10, 20]
})
A1
```

```python
B1 = pd.DataFrame({
    'Names': ['apple', 'graips', 'banana', 'berry'],
    'Price': [8000, 20_000, 24_000, 18_000]
})
B1
```

```python
C2 = pd.merge(B1, A1, left_on = 'Names', right_on = 'Fruit_name', how= 'inner' )
C2
```

```python
C2.drop('Fruit_name', axis = 1, inplace = True)
C2
```

# .pivot() method
# making a dataset generalized by grouping similar features or values

```python
df1 = pd.read_csv('https://raw.githubusercontent.com/MansurCompAI/Practicum_Datasets/refs/heads/master/data_pivot.csv')
df1
```

```python
df2 = df1.pivot(index = 'date', columns = 'item', values = 'value')
df2
```
